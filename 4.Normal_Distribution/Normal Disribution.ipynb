### Answer-1 : expected daily rate of return (gain)

### Answer -2 : stock that have higher risk

```python
sn.distplot(glaxo_df.gain, label = 'Glaxo');
plt.xlabel('gain');
plt.ylabel('Density');
plt.legend();
```

```python
sn.distplot(beml_df.gain, label = 'BEML');
plt.xlabel('gain');
plt.ylabel('Density');
plt.legend();
```

```python
print('Mean:', round(glaxo_df.gain.mean(), 4))
print('Standard Deviation: ', round(glaxo_df.gain.std(), 4))
```

```python
print('Mean: ', round(beml_df.gain.mean(), 4))
print('Standard Deviation: ', round(beml_df.gain.std(), 4))
```

##### Here variance or standard deviation of gain indicates risk. 
* from the above analysis, BEML stock has a higher risk as standard deviation of BEML is 2.64% whereas the standard deviation for Glaxo is 1.33%.


```python
from scipy.stats import norm

#calculate probability that random value is less than 1.06

norm.cdf(1.06)
```

```python
#calculate probability that random value is greater than 1.06

1 - norm.cdf(1.06)
```

```python
from scipy import stats
import scipy

#Probability of making 2% loss or higher in Glaxo

stats.norm.cdf(-0.02, loc = glaxo_df.gain.mean(), scale = glaxo_df.gain.std() )
```

```python
#Probability of making 2% gain or higher in Glaxo
1 - stats.norm.cdf(0.02,loc=glaxo_df.gain.mean(), scale=glaxo_df.gain.std())
```
