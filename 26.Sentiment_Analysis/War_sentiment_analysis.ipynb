```python
stemmer_func = nltk.stem.snowball.SnowballStemmer("english").stem
```

```python
nltk.download('stopwords')
stop = set(stopwords.words('english'))
```

```python
## text cleaning fun.

def pre_process(text):
    text = str(text).lower()
    text = re.sub(r"[-()\"#/@;:{}`+=~|.!?,'0-9]", "", text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stop]
    text=" ".join(text)
    text = [stemmer_func(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
```

```python
war_data['clean_tweets'] = war_data['tweet'].apply(lambda x: pre_process(x))
```

```python
#war_data['token_tweets'] = nltk.word_tokenize(war_data['clean_tweets'])

war_data['token_tweets'] = war_data['clean_tweets'].apply(word_tokenize)
```

```python
def lemma(text):
    lemma_token = " "
    for word in text:
        lemma = word
        lemma_token = lemma_token + " " + lemma
    return lemma_token
```

```python
war_data['Lemma_tweets'] = war_data['token_tweets'].apply(lemma)
war_data.head()
```

```python
def sentiment_analysis(polarity):
    if polarity < 0:
        return 'Negative'
    elif polarity == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

## sentiment analysis using VADER

```python
nltk.download('vader_lexicon')

sentiments = SentimentIntensityAnalyzer()

war_data["vadar_Positive"] = [sentiments.polarity_scores(i)["pos"] for i in war_data["Lemma_tweets"]]
war_data["vadar_Negative"] = [sentiments.polarity_scores(i)["neg"] for i in war_data["Lemma_tweets"]]
war_data["vadar_Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in war_data["Lemma_tweets"]]
war_data["vadar_Compound"] = [sentiments.polarity_scores(i)["compound"] for i in war_data["Lemma_tweets"]]
```

```python
war_data[['tweet', 'textblob_sentiment_polarity','sentiment_textblob', 'vadar_Positive', 'vadar_Negative', 'vadar_Neutral', 'vadar_Compound']].head(20)
```

```python
# function to analyse vadar sentiments
def vader_analysis(compound):
    if compound >= 0.5:
        return 'Positive'
    elif compound <= -0.5 :
        return 'Negative'
    else:
        return 'Neutral'
```

```python
war_data['sentiments_vadar'] = war_data['vadar_Compound'].apply(vader_analysis)
war_data.head()
```
