## Univariate → “one variable” data visualization

```python
#strip plot

sns.stripplot(y= 'tip', data=tips, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

##### Inference: This column contains outliers

```python
#tips.head(25)
```

```python
#Grouping with stripplot()
sns.stripplot(x='day', y='tip', data=tips, jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot

sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
    
```

```python
#tips.head(2)
```

```python
tips.time.unique()
```

```python
tips['day'].unique()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='time')
plt.ylabel('tip ($)')
plt.show()
```

##### Inference: Ppl are preferring dinners on weekends than lunch and vice versa

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

### Box and Violin plot

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')


plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
##Combining plots

sns.violinplot(x='day', y='tip', data=tips, inner=None,color='lightgray')
sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

## Bivariate → “two variables” data visualization

```python
sns.scatterplot(tips['total_bill'], tips['tip'], hue=tips['time'])

# Inference: The tip amount is increasing with the total_bill amount
```

```python
#Joint plot

sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')
plt.show()
```

```python
# Pair plot

sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```
